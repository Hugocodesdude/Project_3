# Function: Company Finder based on Millions/Billions

```python
def millionbillion_company_finder(x):
    
    filter_1 = {"total_money_raised": {"$regex": "M|B"}}
    filter_city_1 = {"offices.0.city": x }
    
    city_conditions = [filter_1,filter_city_1]
    city_finder_list = list(c.find({"$and": city_conditions}, {"_id":0, "name":1,"offices":1}))
    return city_finder_list
```

## Barcelona Million&Billion Dollar Companies

```python
#Barcelona Million Dollar Companies 
millionbillion_company_finder("Barcelona")
```

```python
#Convert into DataFrame
bcn_highval_companies = pd.DataFrame(millionbillion_company_finder("Barcelona"))
bcn_highval_companies
```

```python
#Explode "Offices"
bcn_highval_companies2 = bcn_highval_companies.explode("offices")
bcn_highval_companies2
```

```python
# Concatenate DataFrame
bcn_highval_companies3 = bcn_highval_companies2["offices"].apply(pd.Series)
bcn_highval_companies4 = pd.concat([bcn_highval_companies,bcn_highval_companies3], axis=1)
bcn_highval_companies4
```

```python
#Filter for city:Barcelona
bcn_highval_companies_clean = bcn_highval_companies4[bcn_highval_companies4["city"] =='Barcelona']
bcn_highval_companies_clean
```

```python
# Drop Nans subset=['latitude', 'longitude'])
bcn_highval_companies_clean = bcn_highval_companies_clean.dropna(subset=['latitude', 'longitude'])
```

```python
bcn_highval_companies_clean.to_csv("../data/bcn_highval_companies_clean.csv", index=False)
```
